```python
def check_new_line(ch):
       if ch.isdigit():
          return True

       elif ch.isalpha() and ch!='\n':
         return True
       elif ch == '[' or ch == ']' or ch == '{' or ch =='}' or  ch == '(' or ch == ')' or ch=='"':
          return True
       else:
           return False
check_new_line('"')
```
